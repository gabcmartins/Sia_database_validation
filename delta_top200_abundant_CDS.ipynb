```python
import pandas as pd
```

```python
df= pd.read_csv("/home/gabrielacastilho/Jo/idxstats/Jo_dados_idxstats_com_grupo.tsv", sep= '\t')
```

```python
df_filt = df[df["RPKM"] > 0]
```

```python
print("Linhas antes:", df.shape[0])
print("Linhas depois (RPKM > 0):", df_filt.shape[0])
print("Removidas:", df.shape[0] - df_filt.shape[0])
```

```python
presenca = (
    df_filt
    .groupby(["CDS", "Grupo"])["Amostra"]
    .nunique()
    .unstack(fill_value=0)
)
```

```python
cds_validos = presenca[
    (presenca["Control"] >= 10) &
    (presenca["Parkinson"] >= 10)
].index
```

```python
df_filt = df_filt[df_filt["CDS"].isin(cds_validos)]
```

```python
medianas = (
    df_filt
    .groupby(["CDS", "Grupo"])["RPKM"]
    .median()
    .unstack()
)
```

```python
medianas["delta_PD_vs_Control"] = (
    medianas["Parkinson"] - medianas["Control"]
)
```

```python
medianas = medianas.sort_values(
    "delta_PD_vs_Control",
    ascending=False
)
```

```python
medianas.to_csv(
    "/home/gabrielacastilho/boktor/CDS_delta_PD_vs_Control.tsv",
    sep="\t"
)

#With this file I manually selected the 100 more positive and the 100 more negative 
```
