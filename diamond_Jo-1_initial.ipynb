```python
for arquivo in arquivos:
    # Lê o arquivo e mantém somente linhas com pident >= 40)
    df = pd.read_csv(arquivo, sep="\t")
    dfpident= df[df['pident'] >= 40]
    # Salva sobrescrevendo
    dfpident.to_csv(arquivo, sep="\t", index=False)
```

```python
for arquivo in arquivos:
    # Lê o arquivo
    df = pd.read_csv(arquivo, sep='\t')

    # Conta o total antes do filtro
    total = len(df)

    # Aplica o filtro (evalue <= 1e-4)
    df_filtrado = df[df['evalue'] <= 1e-4]

    # Conta quantas linhas foram excluídas
    excluidas = total - len(df_filtrado)

    # Salva sobrescrevendo o arquivo original
    df_filtrado.to_csv(arquivo, sep='\t', index=False)

    # Mostra o resultado
    print(f"{arquivo}: {excluidas} linhas excluídas")
```

```python
for arquivo in arquivos:
    # Lê o arquivo
    df = pd.read_csv(arquivo, sep='\t')

    # Cria a nova coluna 'cobertura' (length / qlen)
    df['cobertura'] = df['length'] / df['qlen']

    # Salva sobrescrevendo o arquivo original
    df.to_csv(arquivo, sep='\t', index=False)
```

```python
for arquivo in arquivos:
    # Lê o arquivo
    df = pd.read_csv(arquivo, sep='\t')

    # Filtra: mantém só onde cobertura é >=0.5
    df_filtrado = df[df['cobertura'] >= 0.5]

    # Salva sobrescrevendo o arquivo original
    df_filtrado.to_csv(arquivo, sep='\t', index=False)
```

```python
for arquivo in arquivos:
    # Lê o arquivo
    df = pd.read_csv(arquivo, sep='\t')

    # Aplica o filtro (bitscore>=50)
    df_filtrado = df[df['bitscore'] >= 50]

    # Salva sobrescrevendo o arquivo original
    df_filtrado.to_csv(arquivo, sep='\t', index=False)
```

```python
for arquivo in arquivos:
    # Lê o arquivo
    df = pd.read_csv(arquivo, sep='\t')

    # Cria a nova coluna 'qlen/slen'
    df['qlen/slen'] = df['qlen'] / df['slen']

    # Salva sobrescrevendo o arquivo original
    df.to_csv(arquivo, sep='\t', index=False)
```

```python
for arquivo in arquivos:
    # Lê o arquivo
    df = pd.read_csv(arquivo, sep='\t')
    total = len(df)

    # Aplica o filtro 
    df_filtrado = df[(df['qlen/slen'] >= 0.6) & (df['qlen/slen'] <= 2.0)]
    excluidas = total - len(df_filtrado)

    # Salva sobrescrevendo o arquivo original
    df_filtrado.to_csv(arquivo, sep='\t', index=False)
    
    # Mostra o resultado
    print(f"{arquivo}: {excluidas} linhas excluídas")
```

```python
import os

for arquivo in arquivos:
    df= pd.read_csv(arquivo, sep='\t')
    # Ordena pelo bitscore em ordem decrescente
    df_sorted = df.sort_values(by=["qseqid", "bitscore"], ascending=[True, False])

    # Seleciona o melhor hit (primeiro após ordenação) para cada CDS
    best_hits = df_sorted.drop_duplicates(subset="qseqid", keep="first")

    # Gera nome do novo arquivo
    nome_base = os.path.basename(arquivo).replace(".tsv", "")
    novo_nome = f"/home/gabrielacastilho/Jo/diamond_1_inital/{nome_base}_best_hits.tsv"

    # Salva o resultado
    best_hits.to_csv(novo_nome, sep="\t", index=False, header=True)

    print(f"Arquivo salvo: {novo_nome}")
```
